## The transpose method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates =["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True)
bigmac.head(3)
```

## The `swaplevel` method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates =["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True)
bigmac.head(3)
```

```python
bigmac.swaplevel()
```

```python
# The order doesn't matter
bigmac.swaplevel("Date", "Country")
bigmac.swaplevel("Country", "Date")
```

```python
bigmac=bigmac.swaplevel(1, 0)
```

```python
bigmac.head(3)
```

## The .unstack() method, part1

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
world.head(3)
```

```python
s = world.stack()
s.head(3)
```

```python
s.unstack().unstack()
```

```python
s.unstack().unstack().unstack()
```

## The .unstack() method, part2

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
world.head(3)
```

```python
s = world.stack()
s.head(3)
```

```python
s.unstack(-1)
s.unstack(2)


s.unstack(-2)
s.unstack(1)

s.unstack(-3)
s.unstack(0)
```

```python
s.unstack("country")
```

### The .unstack() method, part3

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
s=world.stack()
s.head(3)
```

```python
s.unstack(level = ["country", "year"])
```

```python
s = s.unstack("year", fill_value=0)
```

```python
s.head()
```

## The pivot method

```python
sales = pd.read_csv("salesmen.csv", parse_dates=["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
sales.head(3)
```

```python
sales.pivot(index ="Date", columns="Salesman", values="Revenue")
```

## The pivot_table() method

```python
foods =pd.read_csv("foods.csv")
foods.head(3)
```

```python
foods.pivot_table(values="Spend", index =["Gender", "Item"], columns = "City", aggfunc = "count")
```

```python
pd.pivot_table(data=foods, values="Spend", index =["Gender", "Item"], columns = "City", aggfunc = "count")
```
