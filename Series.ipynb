## The `.get()` Method on a `Series`

```python
pokemon =pd.read_csv("pokemon.csv", usecols = ["Name", "Type 1"], index_col= "Name", squeeze =True)
pokemon.sort_index(inplace=True)
pokemon.head(3)
```

```python
pokemon.get(0)
pokemon.get("Metapod")
```

```python
pokemon.get([0, 5])
pokemon.get(["Meowth", "Moltres"])
```

```python
pokemon.get(10000)
pokemon.get("Nonsense")
pokemon.get([0, 5, 10000])
pokemon.get(["Meowth", "Moltres", "Nonsense"])
```

```python
pokemon.get(key = "Digimon", default="This is not a pokemon")
```

```python
pokemon.get(key = "Charizard", default="This is not a pokemon")
```

## The .apply() method()

```python
google = pd.read_csv("google_stocks.csv", usecols=["Close"], squeeze=True)
google.head(6)
```

```python
def classify_performance(number):
    if number < 300:
        return "OK"
    elif number >= 300 and number < 650:
        return "Satisfactory"
    else:
        return "Incredible"
```

```python
google.apply(classify_performance).tail()
```

```python
google.apply(lambda stock_price : stock_price + 1)
```

## The .map() method()

```python
pokemon_names =pd.read_csv("pokemon.csv", usecols = ["Name"],  squeeze =True)
pokemon_names.head(3)
```

```python
pokemon_types =pd.read_csv("pokemon.csv", usecols = ["Name", "Type 1"], index_col= "Name", squeeze =True)
pokemon_types.head(3)
```

```python
pokemon_names.map(pokemon_types)
```

```python
pokemon_names =pd.read_csv("pokemon.csv", usecols = ["Name"],  squeeze =True)
pokemon_types =pd.read_csv("pokemon.csv", usecols = ["Name", "Type 1"], index_col= "Name", squeeze =True).to_dict()
```

```python
pokemon_names.head()
```

```python
pokemon_types
```
